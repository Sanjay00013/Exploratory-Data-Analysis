While a line plot would show trends over time, a countplot helps me focus on the distribution of attack counts across discrete time periods, making it easier to identify any outliers or significant variations in attack frequencies across different years."

#### Chart - 2

````python
v = data['Region'].value_counts()
dataset = pd.DataFrame({'Region': v.index, 'Count': v.values})
plt.figure(figsize=(16,7))
sns.set(style = 'darkgrid')
plt.xticks(rotation  = 90)
ax=sns.barplot(x='Region',y='Count',data=dataset,palette='coolwarm')
total=float(len(data['Region']))
for p in ax.patches:
  percentage='{:.1f}%'.format(100* p.get_height()/total)
  x=p.get_x()+p.get_width()
  y=p.get_height()
  ax.annotate(percentage,(x,y),ha="center")
plt.title("Region Wise attack (1970-2017)",fontsize=20)
plt.xlabel("Region Name",fontsize=15)
plt.ylabel("Counts",fontsize=15)
plt.show()
````

##### 1. Why did you pick the specific chart?

Answer Here."I chose a bar plot to visualize region-wise attacks and their counts because it effectively compares the frequency of attacks across different regions. Bar plots are ideal for showing categorical data like regions and their corresponding numerical values (counts of attacks). This type of chart allows for easy comparison between regions, highlighting which regions experience higher or lower numbers of attacks. It provides a clear visual representation that helps in identifying patterns and variations in terrorist activities across different geographical areas."
